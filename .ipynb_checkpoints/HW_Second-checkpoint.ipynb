## Split Dataset into Training and Test

_(1.1) Split the dataset into two datasets: 70% training and 30% test. Keep the test set aside_

The rows of our dataset have now all been shuffled. Next want to split the dataset into two datsets: 70% training and 30% test

We are trying to develop a predictive model (predict customer churn). We must have a way to access its accuracy, reliability and credibility. Splitting our dataset into training and test facilitates this. When we say that we want to assess the quality of a predictive model we are saying we would like to know what will happen if we commit to the use of a model in making predictions. Will our predictions be relatively close to the actual outcomes we eventually get?

The simplest way for us to get a handle on the ability of a predictive model to perform on future data is to try to simulate this eventuality. Although we cannot literally gain access to the future before it occurs, we can reserve some of our currently available data and treat it as if were data from the future. We assume the data we have today is a reasonable representation of the data we expect to have in the future. Holding back some of todayâ€™s data for testing is therefore a fair approximation to having future data for testing

### MarriageStatus/Churn

```python
# Find unique values in chosen feature
marriageStatus = pd.unique(df_train.marriageStatus.ravel())

# Insert a new column called 'percent' and fill it with 0s
df_train['percent'] = 0

# Iterate through the unique values in chosen feature and for each value count the amount of customers who churned
# Find the indexes of each row with a particular chosen feature status 
# For each of these row insert count * 100 in the percent column
for p in marriageStatus:
    count = 1 / df_train[df_train.marriageStatus == p].count()['churn']
    index_list = df_train[df_train['marriageStatus'] == p].index.tolist()
    for i in index_list:
        df_train.loc[i, 'percent'] = count * 100
        
# Group dataframe by chosen feature and Churned and sum percent
group = df_train[['percent','marriageStatus','churn']].groupby(['marriageStatus','churn']).sum()

# Plot values of group in a stacked bar chart
my_plot = group.unstack().plot(kind='bar', stacked=True, title="Customer Churn by Marriage Status", figsize=(15,7))

# Define label colours and text and add to the plot
green_patch = mpatches.Patch(color='green', label='Churned')
red_patch = mpatches.Patch(color='blue', label='Didn\'t churn')
my_plot.legend(handles=[green_patch, red_patch], frameon = True)

# Define x and y labels and min and max values for the y axis
my_plot.set_xlabel("Marriage Status")
my_plot.set_ylabel("% Churn")
my_plot.set_ylim([0,100])
```

## Training a Linear Regression Model to Predict Churn

_(2.1) On the training set, train a linear regression model to predict the target feature Churn, using only the descriptive features selected in exercise (1) above_

As mentioned above, the below descriptive features have been used to train the linear regression model:
-  currentHandsetPrice
-  income
-  smartPhone
-  creditRating

```python
# Import statsmodels package for training a linear regression model.
import statsmodels.formula.api as sm

# The formula specifies the target feature and the descriptive features used for training the linear regression model.
# For training the model we call the method fit() on the given data stored in our df dataframe.
lm = sm.ols(formula="churn ~  currentHandsetPrice + income + smartPhone + creditRating", data=df_train).fit()
```

_(2.2) Print the coefficients learned by the model and discuss their role in the model (e.g., interpret the model)_

```python
print(lm.summary())
```

The summary table includes several metrics for evaluating our model. It provides a snapshot of how well this model approximates the relationship between the provided descriptive features (currentHandsetPrice, income, smartPhone, creditRating) and the target feature (churn)

A linear regression is written as:
-  predicted_target = w_0 + w_1 *feature_1 + w_2*feature_2 +â€¦+ w_n*feature_n

w_0 represents the intercept (0.6541). The intercept is the value we would predict for churn if the weight of all the descriptive features was 0 i.e. if the weight for currentHandsetPrice, income, etc. was all 0, then we would predict 0.6541. w_1 to w_n represents the weight for each specific descriptive feature in our model

We will examine the coefficient (column 'coef') of some descriptive features. The coefficient represents the weight that each descriptive feature has on the model. A negative coefficient means that the descriptive feature will give a lower predicted churn value. Conversely, a positive coefficient means that the descriptive feature will give a higher predicted churn value

Customers who fall into the 7th income bracket (-0.1092) and those with a rating of G (-0.1476) will have a lower predicted churn value as these coefficients are negative. It is important to make a distinction for continuous features. income, smartPhone and creditRating are all categorical features. This means that customers can be in only one category for each feature, so the coefficient is multiplied by 1. However, currentHandsetPrice is continuous. This means that for every \$ increase in price, the predicted value of churn drops by 0.0003. This means that although the coefficient for currentHandsetPrice is small (-0.00003), if the price of the phone is high then this can have a big impact on the predicted value for churn. For example, a phone that costs $1,000 will reduce the predicted value of churn by 0.3 (1,000 * -0.0003). These learned coefficients (w_0, w_1, etc.) for each descriptive feature allow us to make predictions for the churn value

_Please Note: This is our first interaction with the classification evaluation measures. For this reason, I have included a brief description of them below. This will not be repeated elsewhere in the notebook. Going forward, I will just focus on the figures themselves_

**Accuracy**<br>
Accuracy measures the ratio of correct predictions to the total number of cases evaluated. Accuracy is a measurement of whether our data is correct. The accuracy for the training set was 0.58. This means that what our model was ~57% accurate. Accuracy is a good starting point, but it is insufficient to use it by itself to determine the efficiency of our predictive model. High accuracy is desirable as it means our model is often correct but it is only a great measure when you have symmetric datasets where values of false positive and false negatives are almost the same
-  Accuracy = TP+TN/TP+FP+FN+TN

**Confusion Matrix**

|  | Predicted No | Predicted Yes |
| --- | --- | --- |
| Actual No | TN: 172 | FP: 172 |
| Actual Yes | FN: 127 | TP: 229 |

True Positives (TP) - These are the correctly predicted positive values which means that the value of actual is yes and the value predicted is also yes

True Negatives (TN) - These are the correctly predicted negative values which means that the value of actual is no and value of predicted is also no

False Positives (FP) â€“ When actual is no and predicted is yes

False Negatives (FN) â€“ When actual is yes but predicted is no

A confusion matrix gives us a visual representation of the performance of our model. For the training set, our model correctly predicted churn 229 times. A confusion matrix allows us to quickly see how our model is performing - specifically whether it is predicting the target too feature too often or not enough. For the training set, our model incorrectly predicted a yes for churn more than it incorrectly predicted a no (172 against 127)

**Precision**<br>
Precision is the number of True Positives divided by the number of True Positives and False Positives. Precision can be thought of as a measure of consistency. A low precision can indicate large numbers of False Positives. The average precision score is 0.57 which is only okay. For the training set, 57% of those predicted by the model to churn/not churn were correct. The relatively high number of FPs (24.5%) will bring it down
-  Precision = TP/TP+FP

**Recall**<br>
Recall is the number of True Positives divided by the number of True Positives and False Negatives. A low recall can indicate a large number of False Negatives. Our average recall score for the training set is 0.57. For the training set, 57% of those that should have been predicted to churn were predicted to churn.  Again, quite low and the exact same as precision
-  Recall = TP/TP+FN

**F-1**<br>
Precision and recall counter each other. If you select almost everything, the precision is very low, while the recall is very high; if you select almost nothing, precision is very high, while the recall is very low. Therefore, the goal is to have some sort of a balance between the two. The most common way to do this is to use the F-1 score. F-1 Score is the weighted average of Precision and Recall. Therefore, this score takes both false positives and false negatives into account. F-1 can often be more useful than accuracy for evaluating our model. Yet again, the average F-1 score is the same: 0.57

**MSE**<br>
Generally, the lower the MSE the higher the accuracy of prediction as this implies a strong match between the actual and predicted data set. The lower the MSE score, the smaller the error and the better our predictor works. However, a low MSE can also be caused by over refinement, where the model is overfitted to the training set. A score of 0.2439 is relatively low and suggests our model is moderately accurate

Start by reading in the full shuffled dataset

```python
df_full = pd.read_csv('FinalUpdatedCustomerChurn-17201032-Shuffle.csv')
```

```python
# change income to category as it goes from 0-9 and doesn't make sense for it to be continuous
df_full['income'] = df_full['income'].astype('category')
```

```python
# convert selected columns to type 'category'
for column in object_columns3:
    df_full[column] = df_full[column].astype('category')
```

```python
# Change churn to an int so we can perform evaluation later
df_full['churn'] = df_full['churn'].astype('int64')
```

We now need to split the shuffled dataset into 5 equal sized subsets. Of the 5 subsets, a single subsample is retained as the validation data for testing the model, and the remaining 4 subsamples are used as training data. The cross-validation process is then repeated 5 times, with each of the 5 subsamples used exactly once as the validation data. The advantage of this method is that all observations are used for both training and validation, and each observation is used for validation exactly once. The 5 results from the different cross-validations can then be averaged to produce a single estimation

```python
# We want to perform a 5-fold cross validation evaluation
# To do this we need to split the dataset into 5 separate subsets
df_full_split = np.split(df_full, 5, axis=0)
df_crosstest1 = df_full_split[0]
df_crosstest2 = df_full_split[1]
df_crosstest3 = df_full_split[2]
df_crosstest4 = df_full_split[3]
df_crosstest5 = df_full_split[4]
df_crosstrain1 = pd.concat([df_full_split[1],df_full_split[2],df_full_split[3],df_full_split[4]], axis=0, join='outer', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)
df_crosstrain2 = pd.concat([df_full_split[0],df_full_split[2],df_full_split[3],df_full_split[4]], axis=0, join='outer', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)
df_crosstrain3 = pd.concat([df_full_split[1],df_full_split[0],df_full_split[3],df_full_split[4]], axis=0, join='outer', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)
df_crosstrain4 = pd.concat([df_full_split[1],df_full_split[2],df_full_split[0],df_full_split[4]], axis=0, join='outer', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)
df_crosstrain5 = pd.concat([df_full_split[1],df_full_split[2],df_full_split[3],df_full_split[0]], axis=0, join='outer', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)
```

### Classification Evaluation Measures for Linear Regression (Cross-Validation) 

```python
crosstrain_list = [df_crosstrain1, df_crosstrain2, df_crosstrain3, df_crosstrain4, df_crosstrain5]
crosstest_list = [df_crosstest1, df_crosstest2, df_crosstest3, df_crosstest4, df_crosstest5]

for i in range (0,5):
    crosstrain = crosstrain_list[i]
    crosstest = crosstest_list[i]
    crosslm = sm.ols(formula="churn ~  currentHandsetPrice + income + smartPhone + creditRating", data=crosstrain).fit()
    crosstest_predict = crosslm.predict(crosstest)
    crossChurnClass = (crosstest_predict > 0.5) * 1
    df_crossChurnClass = pd.DataFrame({'ChurnClass': crossChurnClass})
    y = crosstest['churn']
    print("Test", i+1, "Classification Measurements\n")
    print("Accuracy: ", metrics.accuracy_score(y,df_crossChurnClass))
    print("Confusion matrix: \n", metrics.confusion_matrix(y, df_crossChurnClass))
    print("Classification report:\n ", metrics.classification_report(y, df_crossChurnClass))
    mse = ((crosstest.churn - crosstest_predict)** 2).mean()
    print("\nMean Squared Error:\n", mse)
    mae = abs((crosstest.churn - crosstest_predict)).mean()
    print("\nMean Absolute Error:\n", mae, "\n")
```
